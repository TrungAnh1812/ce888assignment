```python
# Loading MNIST data
from __future__ import print_function
import keras
from keras.datasets import mnist
from keras.models import Sequential
from keras.layers import Dense, Dropout, Flatten
from keras.layers import Conv2D, MaxPooling2D
from keras import backend as K

batch_size = 128
num_classes = 10
epochs = 12

# input image dimensions
img_rows, img_cols = 28, 28

# the data, split between train and test sets
(x_train, y_train), (x_test, y_test) = mnist.load_data()
```

```python
import matplotlib.pyplot as plt

# MNIST displaying function
def display_digit(num, x_train):
    #print(y_train[num])
    #label = y_train[num].argmax(axis=0)
    image = x_train[num].reshape([28,28])
    plt.title('Example: %d' % num)
    plt.imshow(image, cmap=plt.get_cmap('gray_r'))
    plt.show()
```

```python
import random

# Creating random lists
list_1 = list(range(0,28))
rdn_list1 = random.sample(list_1,len(list_1))
rdn_list2 = random.sample(list_1,len(list_1))
rdn_list3 = random.sample(list_1,len(list_1))
```

```python
# Creating sample data sets (4):
x1_train, x1_test = x_train.copy(), x_test.copy()
x2_train, x2_test = x_train.copy(), x_test.copy()
x3_train, x3_test = x_train.copy(), x_test.copy()
x4_train, x4_test = x_train.copy(), x_test.copy()
```

```python
# Shuffle the training data sets
for ob in range(0,x_train.shape[0]):
    for i in range(0,len(rdn_list1)):
        for j in range(0,len(rdn_list1)):
            x1_train[i,j] = x_train[rdn_list1.index(i), rdn_list1.index(j)]
            
    for i in range(0,len(rdn_list2)):
        for j in range(0,len(rdn_list2)):
            x2_train[i,j] = x_train[rdn_list2.index(i), rdn_list2.index(j)]
            
    for i in range(0,len(rdn_list3)):
        for j in range(0,len(rdn_list3)):
            x3_train[i,j] = x_train[rdn_list3.index(i), rdn_list3.index(j)]        
            
```

```python
display_digit(0,x_train)
display_digit(0,x1_train)
display_digit(0,x2_train)
display_digit(0,x3_train)
```

```python
# Shuffle the testing data sets
for ob in range(0,x_test.shape[0]):
    for i in range(0,len(rdn_list1)):
        for j in range(0,len(rdn_list1)):
            x1_test[i,j] = x_test[rdn_list1.index(i), rdn_list1.index(j)]
            
    for i in range(0,len(rdn_list2)):
        for j in range(0,len(rdn_list2)):
            x2_test[i,j] = x_test[rdn_list2.index(i), rdn_list2.index(j)]
            
    for i in range(0,len(rdn_list3)):
        for j in range(0,len(rdn_list3)):
            x3_test[i,j] = x_test[rdn_list3.index(i), rdn_list3.index(j)]        
            
```

```python
display_digit(0,x_test)
display_digit(0,x1_test)
display_digit(0,x2_test)
display_digit(0,x3_test)
```

```python
from keras.datasets import mnist 
from keras.models import Model 
from keras.layers import Input, Dense 
from keras.utils import np_utils 
import numpy as np
from tensorflow.python.ops.variables import trainable_variables

num_train = 60000
num_test = 10000

height, width, depth = 28, 28, 1 # MNIST images are 28x28
num_classes = 10 # there are 10 classes (1 per digit)

(X_train, y_train), (X_test, y_test) = (x1_train, y_train), (x1_test, y_test)

X_train = X_train.reshape(num_train, height * width)
X_test = X_test.reshape(num_test, height * width)
X_train = X_train.astype('float32') 
X_test = X_test.astype('float32')

X_train /= 255 # Normalise data to [0, 1] range
X_test /= 255 # Normalise data to [0, 1] range

Y_train = np_utils.to_categorical(y_train, num_classes) # One-hot encode the labels
Y_test = np_utils.to_categorical(y_test, num_classes) # One-hot encode the labels

input_img = Input(shape=(height * width,))

x = Dense(height * width, activation='relu')(input_img)

encoded1 = Dense(height * width//2, activation='relu')(x)
encoded2 = Dense(height * width//8, activation='relu')(encoded1)

y = Dense(height * width//256, activation='relu')(encoded2)

decoded = Dense(height * width//8, activation='relu')(y)
decoded = Dense(height * width//2, activation='relu')(decoded)

z = Dense(height * width, activation='sigmoid')(decoded)
model = Model(input_img, z)

out = Dense(num_classes, activation='softmax')(y)
reduced = Model(input_img, out)


model.compile(optimizer='adadelta', loss='mse') # reporting the accuracy

#x.trainable=False
#encoded1.trainable=False
#encoded2.trainable=False
#y.trainable=False
reduced.compile(loss='categorical_crossentropy',
          optimizer='adam', 
          metrics=['accuracy']) 


model.fit(X_train, X_train,
      epochs=2,
      batch_size=128,
      shuffle=True,
      validation_data=(X_test, X_test))

#for layer in model.layers:
#weights = model.layers[1].get_weights() # list of numpy arrays
#print(weights)
    
mid = Model(input_img, y)
reduced_representation =mid.predict(X_test)

reduced.fit(X_train, Y_train,
      epochs=2,
      batch_size=128,
      shuffle=True,
      validation_data=(X_test, Y_test))

#weights = model.layers[1].get_weights() # list of numpy arrays
#print(weights)


scores = reduced.evaluate(X_test, Y_test, verbose=1) 
print("Accuracy: ", scores[1])
 
```
